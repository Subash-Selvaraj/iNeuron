```python
df2_date = pd.to_datetime(df2['Date'])
#store the new column for date
df2['converted_date'] = df2_date
#get the month,year,week from the changed_date column ans store as a new column
df2['converted_year'] = df2['converted_date'].dt.year
df2['converted_month'] = df2['converted_date'].dt.month
df2['converted_week'] = df2['converted_date'].dt.isocalendar().week
df2
```

```python
#1 .Find out there avarage rating on weekly basis keep this in a mind that they take two days of leave
Avg_rat_week=df2[(df2['Total Chats'] > 0)][['Agent Name','Average Rating','converted_week']].groupby \
   (['Agent Name','converted_week'])['Average Rating'].mean()
Avg_rat_week.head(20)
```

```python
#valdiation by individual person
df2[(df2['Agent Name']=='Aditya Shinde')&(df2['Total Chats']>0)]['Average Rating'].mean()
#df2.drop('weekly_avg',axis=1,inplace=True)
```

```python
#2 .Total working days for each agents 
toal_working_days=df2[(df2['Total Chats'] > 0)][['Agent Name','converted_date']].groupby \
   (['Agent Name'])['converted_date'].count()
toal_working_days
```

```python
#9 . average weekly response time for each agent
df2['Average Response Time'] = pd.to_datetime(df2['Average Response Time'])
Avg_resp_week=df2[(df2['Total Chats'] > 0)].groupby(['Agent Name','converted_week'])['Average Response Time'].mean()
Avg_resp_week.head(20)
```

```python
#10 . average weekely resolution time for each agents
df2['Average Resolution Time'] = pd.to_datetime(df2['Average Resolution Time'])
Avg_resol_week = df2[(df2['Total Chats']>0)].groupby(['Agent Name','converted_week'])['Average Resolution Time'].mean()
Avg_resol_week.head(20)
```

```python
#13 . Total contributation hour for each and every agents weekly basis
df1_date = pd.to_datetime(df1['Date'])
#store the new column for date
df1['converted_date'] = df1_date
#get the month,year,week from the changed_date column ans store as a new column
df1['converted_year'] = df1['converted_date'].dt.year
df1['converted_month'] = df1['converted_date'].dt.month
df1['converted_week'] = df1['converted_date'].dt.isocalendar().week
total_duration = df1.groupby(['Agent','converted_week'])['Duration'].sum()
total_duration
```
